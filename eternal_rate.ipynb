```python
import pandas as pd
import os
```

```python
# ジャンルIDとジャンル名を対応させる辞書を定義
# (データセット作成時に使用したもの)
genres_map = {
    0:   '未設定',
    101: '異世界（恋愛）',
    102: '現実世界（恋愛）',
    201: 'ハイファンタジー',
    202: 'ローファンタジー',
    301: '純文学',
    302: 'ヒューマンドラマ',
    303: '歴史',
    304: '推理',
    305: 'ホラー',
    306: 'アクション',
    307: 'コメディー',
    401: 'VRゲーム',
    402: '宇宙',
    403: '空想科学',
    404: 'パニック',
    9901: '童話',
    9902: '詩',
    9903: 'エッセイ',
    9904: 'リプレイ',
    9999: 'その他',
    9801: 'ノンジャンル'
}

# コーパスファイルのパス
file_path = "dataset/narou_dataset.csv"
```

```python
# ファイルの存在確認
if not os.path.exists(file_path):
    print(f" エラー: ファイル '{file_path}' が見つかりません。")
    print("先にコーパス作成のセルをすべて実行してください。")
else:
    # コーパスを読み込む
    df = pd.read_csv(file_path)
    print(f" データセット '{file_path}' を読み込みました。（総作品数: {len(df)}件）\n")

    #エタり作品数と非エタり作品数を表示
    eternal_count = df['is_eternal'].sum()
    non_eternal_count = len(df) - eternal_count
    print(f" エタり作品数: {eternal_count}件")
    print(f" 非エタり作品数: {non_eternal_count}件\n")

    # 必要な列が存在するか確認
    if '作品ジャンル' not in df.columns or 'is_eternal' not in df.columns:
        print(" エラー: '作品ジャンル' または 'is_eternal' の列が見つかりません。")
    else:
        # ジャンルごとの作品数を計算
        genre_counts = df['作品ジャンル'].value_counts()
        
        # ジャンルごとにエタり率を計算
        stagnation_rate = df.groupby('作品ジャンル')['is_eternal'].mean() * 100
        
        # 結果を結合してデータフレームを作成
        result_df = pd.DataFrame({
            '作品数': genre_counts,
            'エタり率(%)': stagnation_rate
        })
        
        # ジャンルIDをジャンル名に変換
        result_df['ジャンル名'] = result_df.index.map(genres_map)
        
        # エタり率の高い順に並び替え
        result_df = result_df.sort_values('エタり率(%)', ascending=False)

        print("--- 各ジャンルのエタり率（長期未更新作品の割合） ---")


        # 文字の表示幅を計算
        def get_display_length(text):
            length = 0
            for char in text:
                # 半角or全角を判定
                if len(char.encode()) == 1:
                    length += 1
                else:
                    length += 2
            return length

        # 最大表示幅を計算
        max_len = 0
        for genre_name in result_df['ジャンル名']:
            length = get_display_length(genre_name)
            if length > max_len:
                max_len = length

        # 結果を表示
        for index, row in result_df.iterrows():
            genre_name = row['ジャンル名']
            total_count = int(row['作品数'])
            rate = row['エタり率(%)']
            padding = ' ' * (max_len - get_display_length(genre_name) - 1)
            print(f"{genre_name}{padding}\t| エタり率: {rate:>5.2f}% (全{total_count:>6,}作品)")
```
